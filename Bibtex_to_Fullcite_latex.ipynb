```python
import bibtexparser
import random
import string

def randomString(stringLength=10):
    """Generate a random string of fixed length """
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for i in range(stringLength))




with open('/Users/baptistasmokas/Desktop/Shiny_Project/1_extract/Geometric Deep Learning/bib.bib') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)

for item in bib_database.entries:
    item['ID'] = item['ID'] + randomString(14)
    print(item['ID'])

with open('/Users/baptistasmokas/Desktop/Shiny_Project/1_extract/Geometric Deep Learning/bibtex_RANDOM.bib', 'w') as bibtex_file:
    bibtexparser.dump(bib_database, bibtex_file)
```

```python
import bibtexparser


with open('/Users/baptistasmokas/Desktop/Shiny_Project/1_extract/Geometric Deep Learning/bibtex_RANDOM.bib') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)

result = []
for item in bib_database.entries:
    item['ID'] = item['ID'] + randomString(14)
    item['ID'] = '\\wrapcolorbox[myel]{{\\footnotesize    \\fullcite{' + item['ID'] + '}}}'

    result.append(item['ID'])
    
with open('/Users/baptistasmokas/Desktop/Shiny_Project/1_extract/Geometric Deep Learning/forlatex.txt', 'w') as f:
    for i in result:
        f.write(str(i) + '\n')
    
```
